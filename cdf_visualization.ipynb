```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

paper_data = pd.read_csv('data/workid_authorid_institutionid_pubdate24.csv',index_col=0)
ins_pub = paper_data['institution_id'].value_counts()
```

```python
ins_pub = pd.DataFrame(ins_pub)
cdf = ins_pub['count'].cumsum()/ins_pub['count'].sum()
x_value = cdf[cdf >= 0.95].index[0]

matplotlib.rcParams['font.family'] = 'Times New Roman'
matplotlib.rcParams['font.size'] = 25

plt.figure(figsize=(10, 8))
plt.plot(cdf, label='CDF', linewidth=3)
plt.axvline(x=x_value, color='red', linestyle='--', label='95% of Publications', linewidth=3)
plt.xscale('log')
plt.xlabel('Number of Institutions')
plt.ylabel('Cumulative Distribution of Publications')
plt.title('Cumulative Distribution of Publications per Institution')
plt.grid()
plt.legend()
plt.show()
```
